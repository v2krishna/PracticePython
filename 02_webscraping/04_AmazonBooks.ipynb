```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.amazon.com/best-sellers-books-Amazon/zgbs/books/ref=zg_bs_books_home_all?pf_rd_p=c37c53ab-de39-45c8-93ea-0bdb20185583&pf_rd_s=center-5&pf_rd_t=2101&pf_rd_i=home&pf_rd_m=ATVPDKIKX0DER&pf_rd_r=BEA8E8KB6R8JN033MJXP&pf_rd_r=BEA8E8KB6R8JN033MJXP&pf_rd_p=c37c53ab-de39-45c8-93ea-0bdb20185583"
```

```python
page = requests.get(url)
soup = BeautifulSoup(page.content,"html.parser")
books = soup.find_all(class_="zg-item-immersion")
books[0]
```

```python
first = books[0]
title = first.find(class_="p13n-sc-truncate p13n-sc-line-clamp-1 p13n-sc-truncate-desktop-type2").get_text().strip()
title
```

```python
author = first.find(class_="a-row a-size-small").get_text()
author
```

```python
rating = first.find(class_="a-icon-row a-spacing-none").get_text().strip().replace('\n','  ').split('  ')[0]
rating
```

```python
review = first.find(class_="a-icon-row a-spacing-none").get_text().strip().replace('\n','  ').split('  ')[-1]
review
```

```python
price = first.find(class_="p13n-sc-price").get_text()
price
```

```python
titles , authors, ratings, reviews , prices = [], [], [], [], []
for first in books:
    if first.find(class_="a-icon-row a-spacing-none") != None:
        titles.append(first.find(class_="p13n-sc-truncate p13n-sc-line-clamp-1 p13n-sc-truncate-desktop-type2").get_text().strip())
        authors.append(first.find(class_="a-row a-size-small").get_text())
        ratings.append(first.find(class_="a-icon-row a-spacing-none").get_text().strip().replace('\n','  ').split('  ')[0])
        reviews.append(first.find(class_="a-icon-row a-spacing-none").get_text().strip().replace('\n','  ').split('  ')[-1])
        prices.append(first.find(class_="p13n-sc-price").get_text())
    else:
        titles.append(first.find(class_="p13n-sc-truncate p13n-sc-line-clamp-1 p13n-sc-truncate-desktop-type2").get_text().strip())
        authors.append(first.find(class_="a-row a-size-small").get_text())
        ratings.append(None)
        reviews.append(None)
        prices.append(first.find(class_="p13n-sc-price").get_text())
# titles
```

```python
len(titles) , len(authors), len(ratings), len(reviews), len(prices)
```

```python
d1 = {'Titles':titles, "Authors": authors, 'Ratings': ratings, 'Reviews':reviews, 'prices':prices}
df = pd.DataFrame(d1)
df
```
